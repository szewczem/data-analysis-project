##### Wyznaczenie brakujących BOROUGH (dzielnic) dla danych LONGITUDE, LATITUDE

```python
import geopandas as gpd
from shapely.geometry import Point
```

```python
boroughs = gpd.read_file("NYC_Borough_Boundary.geojson")
boroughs = boroughs.to_crs(epsg=4326)

print(boroughs.columns)
```

```python
# Filter rows with missing BOROUGH and valid LAT/LON
mask = df['BOROUGH'].isna() & df['LATITUDE'].notna() & df['LONGITUDE'].notna()
df_missing = df[mask].copy()

# Create geometry column
geometry = [Point(xy) for xy in zip(df_missing['LONGITUDE'], df_missing['LATITUDE'])]
gdf_missing = gpd.GeoDataFrame(df_missing, geometry=geometry, crs='EPSG:4326')

# Perform spatial join to assign boroughs
joined = gpd.sjoin(gdf_missing, boroughs[['BoroName', 'geometry']], how='left', predicate='within')

# Convert to uppercase to match existing BOROUGH categories
joined['BoroName'] = joined['BoroName'].str.upper()

# Fill in the BOROUGH values back to the main DataFrame
df.loc[joined.index, 'BOROUGH'] = joined['BoroName']

# Done — preview how many were filled
print(f"Filled {joined['BoroName'].notna().sum()} missing boroughs using coordinates.")
```

Określenie najniebezpieczniejszych czynników wypadków dla każdej dzielnicy NYC

```python
df_factors = df.melt(
    id_vars='BOROUGH',
    value_vars=factor_cols,
    value_name='FACTOR'
)

df_factors['FACTOR'] = df_factors['FACTOR'].str.strip().str.title()

df_factors = df_factors.dropna(subset=['FACTOR'])
df_factors = df_factors[df_factors['FACTOR'] != 'Unspecified']
df_factors
```

```python
# Calculate the component sums
sum_injured = df['NUMBER OF PEDESTRIANS INJURED'] + df['NUMBER OF CYCLIST INJURED'] + df['NUMBER OF MOTORIST INJURED']
sum_killed = df['NUMBER OF PEDESTRIANS KILLED'] + df['NUMBER OF CYCLIST KILLED'] + df['NUMBER OF MOTORIST KILLED']

# Masks for updates
mask_injured_update = (df['NUMBER OF PERSONS INJURED'].isna() | (df['NUMBER OF PERSONS INJURED'] == 0)) & (sum_injured > 0)
mask_killed_update = (df['NUMBER OF PERSONS KILLED'].isna() | (df['NUMBER OF PERSONS KILLED'] == 0)) & (sum_killed > 0)

mask_injured_zero = df['NUMBER OF PERSONS INJURED'].isna() & (sum_injured == 0)
mask_killed_zero = df['NUMBER OF PERSONS KILLED'].isna() & (sum_killed == 0)

# Show how many rows will be updated
print(f"Updated NUMBER OF PERSONS INJURED from sum: {mask_injured_update.sum()}")
print(f"Updated NUMBER OF PERSONS KILLED from sum: {mask_killed_update.sum()}")
print(f"Filled missing NUMBER OF PERSONS INJURED with 0: {mask_injured_zero.sum()}")
print(f"Filled missing NUMBER OF PERSONS KILLED with 0: {mask_killed_zero.sum()}")

# Perform updates
df.loc[mask_injured_update, 'NUMBER OF PERSONS INJURED'] = sum_injured
df.loc[mask_killed_update, 'NUMBER OF PERSONS KILLED'] = sum_killed
df.loc[mask_injured_zero, 'NUMBER OF PERSONS INJURED'] = 0
df.loc[mask_killed_zero, 'NUMBER OF PERSONS KILLED'] = 0
```

```python
# Step 1: Normalize and get unique factors
all_factors = df[factor_cols].stack().str.strip().str.title().dropna()
unique_factors = all_factors.unique()

# Step 2: Create mapping dictionary
factor_id_map = {factor: idx+1 for idx, factor in enumerate(unique_factors)}
print("ID map:", dict(list(factor_id_map.items())[:]))

# Step 3: Function to map a row's factors to IDs
def map_factors_to_ids(row):
    ids = []
    for col in factor_cols:
        value = row[col]
        if pd.notna(value):
            key = value.strip().title()
            if key in factor_id_map:
                ids.append(factor_id_map[key])
    return ids

# Step 4: Apply function to each row
df['FACTOR_IDS'] = df.apply(map_factors_to_ids, axis=1)
```

```python
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import matplotlib.patches as mpatches
from matplotlib.colors import ListedColormap

# Step 1: Get top 3 factors by killed persons for each borough and year
factor_yearly_killed = (
    df_exploded.groupby(["BOROUGH", "YEAR", "FACTOR_NAME"], observed=True)["NUMBER OF PERSONS KILLED"]
    .sum()
    .reset_index()
)

# Remove rows with 0 killed
factor_yearly_killed = factor_yearly_killed[factor_yearly_killed["NUMBER OF PERSONS KILLED"] > 0]

top3_per_borough_year = (
    factor_yearly_killed
    .sort_values(["BOROUGH", "YEAR", "NUMBER OF PERSONS KILLED"], ascending=[True, True, False])
    .groupby(["BOROUGH", "YEAR"], group_keys=False, observed=True)
    .head(3)
)

# Step 2: Get unique years and unique factors
years = sorted(top3_per_borough_year["YEAR"].unique())
unique_factors = pd.unique(top3_per_borough_year["FACTOR_NAME"])

# Step 3: Define warm, bright, distinct colors
custom_colors = [
    # "#ff9896",  # light red
    "#7f7f7f",  # gray (remove if not allowed)
      
    "#bcbd22",  # yellow-green 
    "#aec7e8",  # light blue    
    "#1f77b4",  # muted blue
    "#d1410d",
    "#ffd92f",  # bright yellow
    "#2ca02c",  # green
    "#8c564b",  # dark salmon    
    "#e377c2",  # rose pink (remove if pink not allowed)
    "#c5b0d5",  # light lavender (remove if too purple)
    "#87db32",
    "#a6d854",  # lime green
    "#17becf",  # cyan
    "#fc8d62",  # coral orange    
]

sorted_factors = sorted(unique_factors)
cmap = ListedColormap(custom_colors[:len(sorted_factors)])
color_dict = {factor: cmap(i) for i, factor in enumerate(sorted_factors)}

# Step 4: Plotting
for year in years:
    data_year = top3_per_borough_year[top3_per_borough_year["YEAR"] == year]
    boroughs = data_year["BOROUGH"].unique()

    fig, ax = plt.subplots(figsize=(12, 6))

    x_ticks = []
    x_tick_labels = []
    legend_patches = {}
    width = 0.25  # bar width
    spacing = 0.8  # space between borough groups
    pos = 0
    all_bars = []

    for borough in boroughs:
        borough_data = data_year[data_year["BOROUGH"] == borough]
        borough_data = borough_data.sort_values("NUMBER OF PERSONS KILLED", ascending=False)

        for i, (_, row) in enumerate(borough_data.iterrows()):
            factor = row["FACTOR_NAME"]
            killed = row["NUMBER OF PERSONS KILLED"]

            bar = ax.bar(pos, killed, width, color=color_dict[factor])
            all_bars.append((bar[0], killed))  # Save bar and value for annotation

            if factor not in legend_patches:
                legend_patches[factor] = mpatches.Patch(color=color_dict[factor], label=factor)

            pos += width

        # Add space between boroughs
        x_ticks.append(pos - width * 1.5)
        x_tick_labels.append(borough)
        pos += spacing

    # Add value labels
    for bar_obj, value in all_bars:
        if value > 0:
            ax.annotate(f"{int(value)}",
                        xy=(bar_obj.get_x() + bar_obj.get_width() / 2, value),
                        xytext=(0, 3),
                        textcoords="offset points",
                        ha='center', va='bottom',
                        fontsize=8)

    # Formatting
    ax.set_title(f"Top 3 Deadliest Contributing Factors per Borough – {year}", fontsize=14)
    ax.set_ylim(0, 20)
    ax.set_xticks(x_ticks)
    ax.set_xticklabels(x_tick_labels, rotation=0, fontsize=10)
    ax.set_ylabel("Number of Persons Killed")
    ax.set_xlabel(f"Year: {year}", labelpad=20)

    ax.spines["top"].set_visible(False)
    ax.spines["right"].set_visible(False)
    ax.grid(axis="y", linestyle="--", alpha=0.7)

    # Legend
    handles = list(legend_patches.values())
    ax.legend(handles=handles, title="Contributing Factor", bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=9, title_fontsize=10)

    plt.tight_layout()
    plt.show()
```
